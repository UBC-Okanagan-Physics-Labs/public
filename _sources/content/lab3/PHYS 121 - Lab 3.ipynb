# <div style="background-color:rgba(255, 204, 255, 0.5); text-align:center; vertical-align: middle; padding:40px 0; margin-top:30px"><span style="color:rgba(102, 0, 204, 1);">Part 1 - The Electric Field as a Gradient Field (10 minutes)</span></div>



For the pre-lab, you were introduced to scalar fields and vectors fields by thinking about the terrain around a rock quarry. Recall that if we define the altitude above sea level at a position $(x,y)$ by the function $h(x,y)$, the vector field
<br>

\begin{align}
\mathbf{s}(x,y) = \frac{dh}{dx}\,\hat\imath + \frac{dh}{dy}\,\hat\jmath,
\end{align}
<br>
points in the direction of steepest ascent from the point $(x,y)$ and has a magnitude $\left|\mathbf{s}(x,y)\right|$ equal to the slope at that point along the direction of steepest accent. 

Now, suppose you stand on a hill and release a ball from rest. It should spontaneously begin to roll in the direction of steepest *descent* - that is, it will initially roll in the direction of the vector field $-\mathbf{s}(x,y)$.

Analogously, a positive charge released from rest will move in the direction that the electric potential *decreases* the most rapidly. This is exactly the direction the electric field $\mathbf{E}$ points since, for a two-dimensional electric potential $V(x,y)$
<br>

\begin{align}
\mathbf{E}(x,y) = -\left( \frac{dV}{dx}\,\hat\imath + \frac{dV}{dy}\,\hat\jmath \right)
\end{align}
<br>

That is, when we look at our electrical field maps, one way to interpret them is to think about what would happen to a *positive* charge if released from rest in the field. We can also see this in the Lorentz force law, where the force on a charged particle due to an electric field $\mathbf{E}$ (and in the absence of a magnetic field) is given by
<br>

\begin{align}
\mathbf{F}(x,y) = q\mathbf{E}(x,y) = -q\left( \frac{dV}{dx}\,\hat\imath + \frac{dV}{dy}\,\hat\jmath \right)
\end{align}
<br>

Notice that this equation also tells us that a *negative* charge ($q < 0$) will move in the opposite direction that a positive charge would move.

### <div style="background-color:rgba(255, 255, 102, 0.5); text-align:left; padding:20px 0; margin-top:20px">$\quad$Appendix A &ndash; The Mapping Function...</div>

A function for producing false-colour plots of electric potential and then calculating and displaying the corresponding electric field map.  The $x$ and $y$ position data must be entered in units of centimeters.  The $x$ data should span $0$ to $25\rm\ cm$ and the $y$ data should span $0$ to $20\rm\ cm$.  The electric potential data should be in units of volts and span the range from $0$ to $10\rm\ V$.  The function is called as follows:
```python
Mapping(x, y, V, graphNum = 0, vectorField = True, fig_file_name = 'figure.png')
```
The 'x', 'y', and 'V' inputs are lists or arrays of floats and are all required.  'graphNum' is an optional integer input.  If an integer from 1 to 7 is entered, the approximate positions of the silver-paint electrodes will be shown on the final plot.  If other optional input 'vectorField' is equal to True, then the electric field plot will be shown.  The function returns the a single output (the formatted figure):
```python
fig
```
